```python
import os
```

```python
path = 'C:\\Users\\Jake\\Desktop\\labels'
```

```python
morethan2 = []

for file in os.listdir(path):
    with open(path + '\\' + file, 'rt') as f:
        # f.readline()
        # break
        
        if len(f.readlines()) > 1:
            morethan2.append(path + '\\' + file)
```

```python
for item in morethan2:
    with open(item,'rt') as f:
        lines = f.readlines()
        print(item, lines)
```
